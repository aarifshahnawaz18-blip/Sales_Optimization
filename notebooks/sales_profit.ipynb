```python
import pandas as pd

df=pd.read_csv("../content/Superstore.csv",encoding="latin1")
df.info()



```

```python
import seaborn as sns
import matplotlib.pyplot as plt

#outliers of Sales
plt.figure(figsize=(8,5))
sns.boxplot(x=df['Sales'])
plt.title("Sales Boxplot")
plt.show()

#Outliers of Profit
plt.figure(figsize=(8,5))
sns.boxplot(x=df['Profit'])
plt.title("Profit Boxplot")
plt.show()

```

```python
import matplotlib.pyplot as plt

#Category Wise Sales Behaviour 
category_sales = df.groupby('Category')['Sales'].sum().sort_values(ascending=False)

category_sales.plot(kind='bar')
plt.xlabel("category")
plt.ylabel("Quantity Sales")
plt.title("Sales as category")
plt.xticks(rotation=00)

plt.show()



```

```python
#Sub-Category Wise Sales Behaviour 
subcategory_sales = df.groupby('Sub-Category')['Sales'].sum().sort_values(ascending=False)

subcategory_sales.plot(kind='barh')
plt.xlabel("category")
plt.ylabel("Quantity Sales")
plt.title("Sales as Sub-Category")
plt.xticks(rotation=00)

plt.show()
```

```python
#Region  Wise Sales Behaviour
region_sales = df.groupby('Region')['Sales'].sum()

region_sales.plot(kind='bar')
plt.xlabel("Region")
plt.ylabel("Quantity Sales")
plt.title("Sales as Region")
plt.xticks(rotation=00)

plt.show()
```

```python
#Segment Wise Sales Behaviour
segment_sales = df.groupby('Segment')['Sales'].sum()

segment_sales.plot(kind='bar')
plt.xlabel("Segment")
plt.ylabel("Quantity Sales")
plt.title("Sales as Segment")
plt.xticks(rotation=00)

plt.show()

```

```python
#Yearly Wise Sales Behaviour

month_profit=df.groupby(df['Order Date'].dt.month)['Profit'].sum()


month_profit.plot(kind='bar')
plt.xlabel("Segment")
plt.ylabel("Profit")
plt.title("Profit as Segment")
plt.xticks(rotation=00)

plt.show()

```
